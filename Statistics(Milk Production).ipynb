##### C.1.1.  T-Test, one population

I am interested to test the average Milk produced in Tonnes in Ireland using T-Test. The mean of the amount of milk produced in Ireland for 2011 to 2021 was 7003 Tonnes. I want to verify the mean if it is still the same if I just take into account the most recent 5 years (n=5). I chose 5 because that is the median of the sample population available. At a 5% significance level, I want to test amounts from 2017 to 2021.

```python
DS_Ireland
```

```python
## getting the 5 most recent years data

IE_tail= DS_Ireland.tail(5)
IE_tail
```

```python
IE_tail.shape
```

```python
# x is the amount (in Tonnes) of milk produced and utilised in Ireland

X=IE_tail['Amount']; X
```

The mean value of milk produced and utilised in Ireland for 2011 to 2021 is 7003 Tonnes. My null hypothesis is that 2017-2021 mean amount will also give 7003 Tonnes.

```python
#H0 : u = 7003
#H1 : u =! 7003
#stats.ttest_1samp(X,mu of H0)

stats.ttest_1samp(X,7003)
```

The P-value is less than 0.05, therefore, the null hypothesis is rejected and the alternative hypothesis is accepted. There is enough evidence to say that the mean amount of milk produced in Ireland for 5 years is not equal to 7003 Tonnes.

Now, I want to test if the mean amount of milk produced and utilised in Ireland for 5 years is less than 7003 Tonnes.

```python
#H0 : u < 7003
#H1 : u > 7003
stats.t.sf(4.561763386049194, 4)
```

The p-value I got is less than 0.05, therefore, I reject the null hypothesis. There is enough evidence to say that the mean amount of milk produced and utilised in Ireland is greater than 7003.

```python

test = stats.ttest_ind(Ire, Neth, 
                      equal_var=False)

display(test)
```

The p-value is larger than the threshold (alpha=0.05), then there is evidence against the null hypothesis of equal population means.
Ireland sample mean is not equal to Netherlands sample mean.

#### Linear Regression for Ireland

```python
DS_Ireland_lr = DS_Ireland.drop('Country',axis=1)
DS_Ireland_lr 
```

```python
scaler = StandardScaler()

df_scale = pd.DataFrame(scaler.fit_transform(DS_Ireland_lr ),columns=DS_Ireland_lr.columns)
```

```python
df_scale.head()
```

```python
y=DS_Ireland_lr['Amount']
x = df_scale.drop(['Amount'],axis=1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state= 40)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
lrm = LinearRegression()
lrm.fit(X_train, Y_train) #fit an OLS model
```

```python
y_preds_train = lrm.predict(X_train)
y_preds_test = lrm.predict(X_test)  #making predictions
```

```python
print("R-squared of the model in training set is: {}".format(lrm.score(X_train, Y_train)))
print("-----Test set statistics-----")
print("R-squared of the model in test set is: {}".format(lrm.score(X_test, Y_test)))
print("Root mean squared error of the prediction is: {}".format(rmse(Y_test, y_preds_test)))
print("Mean absolute percentage error of the prediction is: {}".format(np.mean(np.abs((Y_test - y_preds_test) / Y_test)) * 100))
```

A value of 1 indicates that the response variable can be perfectly explained without error by the predictor variable.
